```python
import pandas as pd#CONVERTING ORDER DATES TO THE FORMAT
df5 = pd.read_csv('Sales.2.csv')
df5['Order Date'] = df5['Order Date'].apply(lambda x:pd.to_datetime(x, infer_datetime_format=True))
df5['Order Date'] = df5['Order Date'].dt.strftime('%d-%m-%Y')
print(df5[['Order Date']])
```

```python
df5['Order Date'] = pd.to_datetime(df5['Order Date'], errors='coerce', infer_datetime_format=True)
print("After conversion to datetime:")
print(df5['Order Date'])
df5['Order Date'] = df5['Order Date'].dt.strftime('%d-%m-%Y')
print("\nFormatted Order Date:")
print(df5['Order Date'])
```

```python
import pandas as pd#CONVERTING DELIVERY DATES TO THE FORMAT
df5 = pd.read_csv('Sales.2.csv')
df5['Delivery Date'] = df5['Delivery Date'].apply(lambda x:pd.to_datetime(x, infer_datetime_format=True))
df5['Delivery Date'] = df5['Delivery Date'].dt.strftime('%d-%m-%Y')
print(df5[['Delivery Date']])
```

```python
print(df5['Delivery Date'].head())
```

```python
print(df5['Delivery Date'].dtype)
```

```python
df5['Delivery Date'] = pd.to_datetime(df5['Delivery Date'], errors='coerce', infer_datetime_format=True)
print("After conversion to datetime:")
print(df5['Delivery Date'])
df5['Delivery Date'] = df5['Delivery Date'].dt.strftime('%d-%m-%Y')
print("\nFormatted Delivery Date:")
print(df5['Delivery Date'])
```

```python
#ANALYSING THE MEAN AND MEDIAN OF DELIVERY TIME
df5['Order Date'] = pd.to_datetime(df5['Order Date'], format='%d-%m-%Y', errors='coerce')
df5['Delivery Date'] = pd.to_datetime(df5['Delivery Date'], format='%d-%m-%Y', errors='coerce')
df5['Delivery Time'] = (df5['Delivery Date'] - df5['Order Date']).dt.days
mean_delivery_time = df5['Delivery Time'].mean()
median_delivery_time = df5['Delivery Time'].median()
print(f"Mean Delivery Time: {mean_delivery_time}")
print(f"Median Delivery Time: {median_delivery_time}")
```

```python
#ANALYSING THE MISSING DELIVERY DATE WITH RESPECT TO CURRENCY
missing_count = df5['Delivery Date'].isnull().sum()
missing_percentage = (missing_count / len(df5)) * 100
missing_by_currency = df5[df5['Delivery Date'].isnull()].groupby('Currency Code').size()
print(f"Number of missing values in Delivery Date: {missing_count}")
print(f"Percentage of missing values in Delivery Date: {missing_percentage:.2f}%")
print("Number of missing values by Currency Code:")
print(missing_by_currency)
```
